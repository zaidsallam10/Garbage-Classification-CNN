This is a garbage classification infering code (prediction NOT training), trained on a public dataset from Kaggle using VGG-16 computer vision architecture with >85% accuracy, it can detect or disinguish between 6 lables ('cardboard','glass','metal','paper','plastic','trash').




```python
import numpy as np
np.random.seed(123)  # for reproducibility
import keras

from keras.models import Sequential
from keras.layers import Dense,MaxPooling2D,Conv2D,Flatten, Dropout, Activation, Flatten
# from sklearn.model_selection import train_test_split 
# from sklearn.metrics import classification_report 


import os
import matplotlib.pyplot as plt
import matplotlib.image as img
import tensorflow as tf
from keras.preprocessing import image
from keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.python.client import device_lib
from IPython.display import Image


CATEGORIES=['cardboard','glass','metal','paper','plastic','trash']

```

```python
def evaluate(img_fname):
    img = image.load_img(img_fname)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    preds = model.predict(x)
    pred_name = CATEGORIES[np.argmax(preds)]
    print('predicted =',pred_name)
```

```python
model=tf.keras.models.load_model('/content/drive/MyDrive/best_model_garbage_2021_03_12_2vgg.hdf5')
```
